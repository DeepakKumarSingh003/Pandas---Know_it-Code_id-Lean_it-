* Pandas is an open-source library that provides high-performance data manipulation in Python. 
* The name of Pandas is derived from the word Panel Data,
* Panel data, sometimes referred to as longitudinal data, is data that contains observations about different cross sections across time. Examples of groups that may make up panel data series include countries, firms, individuals, or demographic groups.

* Data analysis requires lots of processing, such as restructuring, cleaning or merging, etc. There are different tools are available for fast data processing, such as Numpy, Scipy, Cython, and Panda. But we prefer Pandas because working with Pandas is fast, simple and more expressive than other tools.

* Pandas is built on top of the Numpy package

* It can perform five significant steps required for processing and analysis of data irrespective of the origin of the data, i.e., load, manipulate, prepare, model, and analyze.

#Key Features of Pandas
* It has a fast and efficient DataFrame object with the default and customized indexing.
* Used for reshaping and pivoting of the data sets.
* Group by data for aggregations and transformations.
* It is used for data alignment and integration of the missing data.
* Provide the functionality of Time Series.
* Process a variety of data sets in different formats like matrix data, tabular heterogeneous, time series.
* Handle multiple operations of the data sets such as subsetting, slicing, filtering, groupBy, re-ordering, and re-shaping.
* It integrates with the other libraries such as SciPy, and scikit-learn.
Provides fast performance, and If you want to speed it, even more, you can use the Cython.

#Benefits of Pandas
* Data Representation: It represents the data in a form that is suited for data analysis through its DataFrame and Series.

* Clear code: The clear API of the Pandas allows you to focus on the core part of the code. So, it provides clear and concise code for the user.

```python
# creating series with list. Self indexing here
list1=['I','Love','Python','because','of','pandas','as','its','library']

myseriesl1=pd.Series(list1)
print(myseriesl1)

```

  * **We can see the indexs from 0 to 8 total 9 elements**


```python
#Accessing Element Using self Label (index) 
print(myseriesl1[:5]) #Give me first 5 elements only.
print("*"*30)
#Give me 3rd element only
print("3rd element from series:",myseriesl1[2])  

#You might be thinking why 2 when need third element as in python indexing start from 0. So count 0th element as 1.
```

* **But then you might say "I dont like this indexing from 0"**
* No Worries we have a solution.......

```python
#.....Create user defined index in Series
#Createing index 
myseries=pd.Series(list1,index=[1,2,3,4,5,6,7,8,9])
print(myseries)
```

* **It is fine now**


```python
# Check out index manually and its type.
myseries.index
```

```python
#Now calling 3rd element
print("3rd element from series:-",myseries[3])  
```

**Create pandas series with a dictionary**

```python
# Create a simple Pandas Series from a dictionary:
dict1={ 'Python':1990, 'Visual Basic':1991 ,'R':1993 , 'Ruby':1995,'Java':1995, 'JavaScript':1995, 'PHP':1995}
myseries_dict=pd.Series(dict1)
print(myseries_dict)
#Give me 5th element of series
print('5th element:-',myseries_dict['Python'])
```

      While using dictionary to create series then Key becomes your index and value become element.

```python
#Operation on Series
myseries_dict[myseries_dict<=1993] #Give me those elements from Series which are less than 1993
```

```python
#Update element
myseries_dict['Python']=None
myseries_dict
```

```python
#Check is there any null value in the series
myseries_dict.isnull()
```

**Create DataFrame using Nested List**

```python
#List of List
nested_list=[["MAE","Mean Absolute Error"],["MSE","Mean Square Error"],["RMSE","Root Mean Square Error"],["R2","Coefficient Of Determination"]]

#Create dataFrame with list of list
my_df2=pd.DataFrame(nested_list)
print(my_df2)
```

    Do you need column names as Evaluation Matrix and Full Form? Let Do that!

```python
# Columns with name and dataframe with list of list
my_df3=pd.DataFrame(nested_list,columns=["Name","Full_Form"]) #Using Columns we can assign names of columns
print(my_df3)

# initialize list of lists
data = [['DS', 'Linked_list', 10], ['DS', 'Stack', 9], ['DS', 'Queue', 7],
        ['Algo', 'Greedy', 8], ['Algo', 'DP', 6], ['Algo', 'BackTrack', 5], ]
 
# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['Category', 'Name', 'Marks'],index=['#1',"#2","#3","#4","#5",'#6'])
 
# print dataframe.
print(df )
```

**Create DataFrame with list of Dictionary**

```python
#DataFrame from list of dictionary
my_list_dict=pd.DataFrame([
    {'Score': 250, 'place': 'Mumbai'},
    {'Score': 180, 'place': 'Ahemedabad'},
    {'Score': 230, 'place': 'Rajkot'},
])
my_list_dict
```

**Create DataFrame with list of tuples**

```python
list_of_tup=[(n,n**2,n**3) for n in range(1,10)]
list_of_tup
my_lot_df=pd.DataFrame(list_of_tup,columns=['Num','Square','Cube'])
my_lot_df
```

**Create DataFrame from pandas Series as value of dictionary**

```python
# Store the data in a dictionary
series_dict = {'First Series' : pd.Series([10, 20, 30, 40]), 
              'Second Series' : pd.Series([10, 20, 30, 40])} 
  
# Create the Dataframe using the dictionary
df = pd.DataFrame(series_dict) 
  
# Print the dataframe
df 
```

**Create DataFrame from Numpy Array**

```python
my_np_df=pd.DataFrame(np.array([[1,1,1],[2,4,8],[3,9,27]]),columns=['Num','Square','Cube'])
my_np_df
```
